```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
daily_data = pd.read_csv('../Data/daily.csv', index_col=0)
```

```python
daily_data
```

```python
daily_X = daily_data.drop(columns=['cnt', 'casual', 'registered'])
daily_y = daily_data.cnt
```

```python
daily_X.corrwith(daily_y)
```

```python
daily_X.corrwith(daily_y).abs().plot.bar()
plt.show()
```
